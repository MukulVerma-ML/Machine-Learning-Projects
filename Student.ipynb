```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score , precision_score , recall_score , f1_score , confusion_matrix
from sklearn.linear_model import LinearRegression , Ridge , Lasso , LogisticRegression
```

```python
df= pd.read_excel('Student_Performance.xlsx')

df['Extracurricular Activities']=df['Extracurricular Activities'].astype(str).str.strip().str.lower().map({'yes':1 , 'no':0})
df.head()
```

```python
#independent_features

#X=df.drop(columns=['Performance Index'])
X=df[['Hours Studied','Extracurricular Activities','Previous Scores','Sleep Hours', 'Sample Question Papers Practiced']]

#dependent_features
Y_marks=df['Performance Index'] #regression  ke liye
Y_pass=(df['Performance Index'] >=40).astype(int) #classification ke liye. >=40+ wala pass and <40 wala fail
```

```python
#train-test split
X_train , X_test , Y_marks_train , Y_marks_test = train_test_split(X,Y_marks,test_size=0.2,random_state=42)
X_train , X_test , Y_pass_train , Y_pass_test = train_test_split(X,Y_pass,test_size=0.2,random_state=42)
```

```python
#feature_scaling
scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
#regression model

lr=LinearRegression()
lr.fit(X_train,Y_marks_train)
Y_marks_pred=lr.predict(X_test)

Y_pred_lr=lr.predict(X_test)
print("Linear Regression R^2 score:",round(r2_score(Y_marks_test,Y_pred_lr)))
```

```python
ridge=Ridge(alpha=5)
ridge.fit(X_train,Y_marks_train)
Y_pred_ridge=ridge.predict(X_test)

print("Ridge Regression R^2 score:",round(r2_score(Y_marks_test,Y_pred_ridge)))
```

```python
lasso=Lasso(alpha=5)
lasso.fit(X_train,Y_marks_train)
Y_pred_lasso=lasso.predict(X_test)

print("Lasso Regression R^2 score:",round(r2_score(Y_marks_test,Y_pred_lasso)))
```

```python
#Classification model

log=LogisticRegression()
log.fit(X_train,Y_pass_train)
Y_pred_pass=log.predict(X_test)

#print("Logistic Regression Accuracy:",round(log.score(X_test,Y_pass_test)*100,2))

print("Precision:",round(precision_score(Y_pass_test,Y_pred_pass),4))
print("Recall:",round(recall_score(Y_pass_test,Y_pred_pass),4))
print("F1 Score:",round(f1_score(Y_pass_test,Y_pred_pass),4))
print("Confusion Matrix:\n",confusion_matrix(Y_pass_test,Y_pred_pass))
```
